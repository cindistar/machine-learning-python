### Linear Regression
An example of how to build a linear model regression for machine learning

![](robot.jpg)

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df = pd.read_csv("homeprices1.csv")
df
```

```python
%matplotlib inline
plt.xlabel("area(sqr ft)")
plt.ylabel("price(US$)")
plt.scatter(df.area, df.price, color="red", marker="+")
```

```python
lg = linear_model.LinearRegression()
```

```python
lg.fit(df[["area"]], df.price)
```

```python
lg.predict([[4300]]) # area
```

```python
lg.coef_ # m
```

```python
lg.intercept_ # b
```
